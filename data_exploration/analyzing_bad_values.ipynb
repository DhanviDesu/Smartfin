## Calculate standard deviation

```python
import statistics
imu = df['IMU A2']
imu = np.array(imu)
std = statistics.stdev(imu)
```

```python
bad_indices = []
for i in range(0, len(imu)):
    if imu[i] - imu.mean() > (std):
        bad_indices.append(i)
        
bad_indices
```

```python
bad_rows = pd.DataFrame()
# for i in bad_indices:
#     bad_rows = bad_rows.append(df.iloc[i]) 
    
bad_rows = df.iloc[bad_indices]
bad_rows = bad_rows.drop(['IMU A1', 'IMU A2', 'IMU A3', 'IMU G1', 'IMU G2', 'IMU G3', 'IMU M1', 'IMU M2', 'IMU M3', 'Time', 'TimeDelta'], axis=1)
```
